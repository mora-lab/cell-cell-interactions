## 1. Data Preparation

```r
# install packages
installed <- installed.packages()[,"Package"]
packages.required = c("Seurat", "dplyr", "patchwork", "Matrix", "parallel","NMF","SingleCellSignalR")

for (package in packages.required){
    if (!(package %in% installed)) {
     #install.packages(package)
     if (!requireNamespace("BiocManager", quietly = TRUE))
         install.packages("BiocManager")
     
     BiocManager::install(package, update = FALSE)
  }
}

if(!require(devtools)) install.packages("devtools")
devtools::install_github("sqjin/CellChat")
devtools::install_github("Coolgenome/iTALK")
devtools::install_github("BaderLab/CCInx")
devtools::install_github("YUZIXD/scMLnet")
```

```r
# load packages
suppressPackageStartupMessages(library(Seurat))
suppressPackageStartupMessages(library(dplyr))
suppressPackageStartupMessages(library(patchwork))
suppressPackageStartupMessages(library(Matrix))
suppressPackageStartupMessages(library(parallel))
suppressPackageStartupMessages(library(iTALK))
suppressPackageStartupMessages(library(CellChat))
suppressPackageStartupMessages(library(SingleCellSignalR))
suppressPackageStartupMessages(library(CCInx))
suppressPackageStartupMessages(library(scMLnet))

# load tool functions
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/run_CellChat.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/run_iTalk.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/run_SingleCellSignalR.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/run_scMLnet.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/run_CCInx.R")

# load single-cell sample data
sample <- readRDS("git@github.com:mora-lab/cell-cell-interactions/blob/main/benchmark-workflow/data/sample.rds")
```

```r
# normalized data of the sample
str(sample@assays$RNA@data)
```

```r
# cell types and amounts of the sample
table(Idents(sample))
```

## 3. SSP Calculation

```r
# load gold standard sample
gold <- read.table("git@github.com:mora-lab/cell-cell-interactions/blob/main/benchmark-workflow/data/Gold Standard (sample).txt", header = TRUE, sep = "\t")

# load SSP calculation functions
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/CellChat_SSP_calculation.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/iTalk_SSP_calculation.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/SingleCellSignalR_SSP_calculation.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/CCInx_SSP_calculation.R")
source("git@github.com:mora-lab/cell-cell-interactions/tree/main/benchmark-workflow/R/scMLnet_SSP_calculation.R")
```

```r
head(gold)
```

```r
CellChat_SSP <- run_CellChat_SSP(CellChat,gold)
iTalk_SSP <- run_iTalk_SSP(iTalk,gold)
SingleCellSignalR_SSP <- run_SingleCellSignalR_SSP(SCSR,gold)
CCInx_SSP <- run_CCInx_SSP(CCInx,gold)
scMLnet_SSP <- run_scMLnet_SSP(scMLnet,gold)
```

```r
SSP_list <- list(CellChat=CellChat_SSP, iTalk=iTalk_SSP, SingleCellSignalR=SingleCellSignalR_SSP, CCInx=CCInx_SSP,scMLnet=scMLnet_SSP)
```
